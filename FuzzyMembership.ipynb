```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def setRange():
    print("Enter range")
    f_range=float(input("Enter lower limit"))
    l_range=float(input("Enter upper limit"))
    num_divisions=int(input("Enter number of divisions"))
    x_values=np.linspace(f_range,l_range,num_divisions)
    return x_values
```

```python
def plotting(x_values,membership_degree,x_label,y_label,title):
    plt.plot(x_values,membership_degree,color='blue')
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.grid(True)
    plt.show()
```

```python
def triangular(x,a,b,c):
    return max(min(((x-a)/(b-a)),((c-x)/(c-b))),0)

x_values=setRange()

x1_value=float(input("Enter value of a"))
x2_value=float(input("Enter value of b"))
x3_value=float(input("Enter value of c"))

membership_degree=[triangular(x,x1_value,x2_value,x3_value) for x in x_values]

plotting(x_values,membership_degree,'x','Degree Of Membership','Triangular Membership Function')
    
```

```python
def trapezoidal(x,a,b,c,d):
    return max(min(((x-a)/(b-a)),1,((d-x)/(d-c))),0)

x_values=setRange()

x1_value=float(input("Enter value of a"))
x2_value=float(input("Enter value of b"))
x3_value=float(input("Enter value of c"))
x4_value=float(input("Enter value of d"))

membership_degree=[trapezoidal(x,x1_value,x2_value,x3_value,x4_value) for x in x_values]

plotting(x_values,membership_degree,'x','Degree Of Membership','Trapezoidal Membership Function')
```

```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

x_values=setRange()

y_values = [sigmoid(x) for x in x_values]

plotting(x_values, y_values, 'x',  'Sigmoid(x)', 'Sigmoid Function')
```

```python
def generalized_bell(x,centre,width,shape):
    return 1/(1+np.abs(((x-centre)/width)**(2*shape)))
              
x_values=setRange()

centre_value=float(input("enter centre value"))
width_value=float(input("enter width value"))
shape_value=float(input("enter shape value"))
              
y_values=generalized_bell(x_values,centre_value,width_value,shape_value)

plotting(x_values,y_values,'X','degree of membership','Generalized Bell Membership Function')
```

```python
def gaussian(x):
    m=np.mean(x)
    std=np.std(x)
    y_out=np.exp(-(x-m)**2/(2 * std**2))
    return y_out
    
x_values=setRange()

y_values=gaussian(x_values)

plotting(x_values,y_values,'x','degree of membership','Gaussian Membership Function')
```
